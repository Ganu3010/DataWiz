### Which poison pokemon has the strongest attack ?

```python
x = ds[((ds['Type 1'] == 'Poison') | (ds['Type 2'] == 'Poison'))]['Attack'] == ds[((ds['Type 1'] == 'Poison') | (ds['Type 2'] == 'Poison'))]['Attack'].max()
ds[((ds['Type 1'] == 'Poison') | (ds['Type 2'] == 'Poison')) & x]
```

###Which Generation has the lowest number of Pokemon?

```python
ds.Generation.value_counts()
```

##Which Generation has the highest number of Legendary Pokemon?

```python
for i in range(1,7):
    g1 = ds.groupby('Generation').get_group(i)
    print(i ,g1[g1['Legendary']].count())
```

###  Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
speeds = list(set(ds[(ds['Type 1'] == 'Ghost') |(ds['Type 2'] == 'Ghost') ].sort_values('Speed')['Speed']))
ds[((ds['Type 1'] == 'Ghost') | (ds['Type 2'] == 'Ghost')) & (ds['Generation'] == 4) & (ds['Speed'] == speeds[-2])]
```

### How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
a = int(ds[ds['Name'] == 'Charizard']['Attack'])
d = int(ds[ds['Name'] == 'Charizard']['Defense'])
len(ds[(ds['Attack']<a) & (ds['Defense']>d) & (ds['Legendary'] == False)])
```

### Which pokemon has the highest combined value of Attack and Sp. Atk ?

```python
ds['tattack'] = ds['Attack'] + ds['Sp. Atk']
ds[ds['tattack'] == ds['tattack'].max()]
```

### Which type of legendary pokemons are the most common?

```python
n = ds[ds['Legendary']]['Type 1'].value_counts().head()
m = ds[ds['Legendary']]['Type 2'].value_counts().head()
print(n)
print(m)
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
min6 = ds[(ds['Legendary']) & (ds['Generation'] == 6)]['HP'].min()
len(ds[(ds['Legendary'] == False) & (ds['Generation'] == 3) & (ds['HP']>min6)])
```

### Print out the third slowest pokemon(s) in the dataset.

```python
speeds = list(set(ds.sort_values('Speed')['Speed']))
ds[ds['Speed'] == speeds[2]]
```

### Which pokemon type has the highest average HP?

```python
x = list(set(ds['Type 1']))
x.sort()
d = {}
for i in x:
    d[i] = ds.groupby('Type 1').get_group(i).HP.mean()
y = list(set(ds['Type 2']))
y.pop(0)
y.sort()
d2 = {}
for i in y:
    d2[i]=ds.groupby('Type 2').get_group(i).HP.mean()
# print(d,d2,sep='\n')
m = max(zip(d.values(), d.keys()))
n = max(zip(d2.values(), d2.keys()))
print(m,n)
```
